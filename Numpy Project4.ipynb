```python
import numpy as np
np.set_printoptions(suppress=True, precision=2)
finance=np.genfromtxt('house_finances.csv', delimiter=',', skip_header=True)
#to find the total expenses of a household
total_expenses=finance[ : ,2:9].sum(axis=1)
finance=np.column_stack((finance,total_expenses))
#to find the amount of savings in the household
savings=finance[ : ,1:2].sum(axis=1)-finance[ : ,2:9].sum(axis=1)
income=finance[ : ,1:2].sum(axis=1)
save_percent=(savings/income)*100
finance=np.column_stack((finance,savings,save_percent))
print(finance)
```

```python
#to find the average income and expenses of each category
avg_finance=finance[ : ,1:9].mean(axis=0)
print("The average income and expenditure in certian categories are:",avg_finance)
```

```python
#to find the highest and lowest income and expenses in each category
high_income = finance[finance[:, 1] > 100000][:, [0, 1]]
low_income = finance[finance[:,1]<50000][:, [0, 1]]
high_mortgage=finance[finance[:,2]>25000][:, [0, 2]]
low_mortgage=finance[finance[:,2]<2500][:, [0, 2]]
high_utilities=finance[finance[:,3]>25000][:, [0, 3]]
low_utilities=finance[finance[:,3]<2500][:, [0, 3]]
high_grocery=finance[finance[:,4]>25000][:, [0, 4]]
low_grocery=finance[finance[:,4]<2500][:, [0, 4]]
high_transport=finance[finance[:,5]>25000][:, [0, 5]]
low_transport=finance[finance[:,5]<2500][:, [0, 5]]
high_insurance=finance[finance[:,6]>25000][:, [0, 6]]
low_insurance=finance[finance[:,6]<2500][:, [0, 6]]
high_healthcare=finance[finance[:,7]>25000][:, [0, 7]]
low_healthcare=finance[finance[:,7]<2500][:, [0, 7]]
high_entertainment=finance[finance[:,8]>25000][:, [0, 8]]
low_entertainment=finance[finance[:,8]<2500][:, [0, 8]]
high_misc=finance[finance[:,9]>25000][:, [0, 9]]
low_misc=finance[finance[:,9]<2500][:, [0, 9]]
high_expenses=finance[finance[:,10]>80000][:, [0, 10]]
low_expenses=finance[finance[:,10]<25000][:, [0, 10]]
high_savings=finance[finance[:,-2]>50000][:, [0, -2]]
low_savings=finance[finance[:,-2]<2500][:, [0, -2]]
high_save_percent=finance[finance[:,-1]>25][:, [0, -1]]
low_save_percent=finance[finance[:,-1]<5][:, [0, -1]]
```

```python
#to print the households with high and low income and expenses in each category
print("The households with high income are:",high_income)
print("The households with low income are:",low_income)
print("The households with high mortgage are:",high_mortgage)
print("The households with low mortgage are:",low_mortgage)
print("The households with high utilities are:",high_utilities)
print("The households with low utilities are:",low_utilities)
print("The households with high grocery expenses are:",high_grocery)
print("The households with low grocery expenses are:",low_grocery)
print("The households with high transport expenses are:",high_transport)
print("The households with low transport expenses are:",low_transport)
print("The households with high insurance expenses are:",high_insurance)
print("The households with low insurance expenses are:",low_insurance)
print("The households with high healthcare expenses are:",high_healthcare)
print("The households with low healthcare expenses are:",low_healthcare)
print("The households with high entertainment expenses are:",high_entertainment)
print("The households with low entertainment expenses are:",low_entertainment)
print("The households with high misc expenses are:",high_misc)
print("The households with low misc expenses are:",low_misc)
print("The households with high total expenses are:",high_expenses)
print("The households with low total expenses are:",low_expenses)
print("The households with high savings are:",high_savings)
print("The households with low savings are:",low_savings)
print("The households with high save percent are:",high_save_percent)
print("The households with low save percent are:",low_save_percent)
```

```python
#to find the correlation between income and expenses
correlation=np.corrcoef(finance[:,1],finance[:,10]) 
print("The correlation between income and expenses is:",correlation[0,1])
#to find the correlation between income and savings
correlation_savings=np.corrcoef(finance[:,1],finance[:,-2])
print("The correlation between income and savings is:",correlation_savings[0,1])
#to find the correlation between income and save percent
correlation_save_percent=np.corrcoef(finance[:,1],finance[:,-1])
print("The correlation between income and save percent is:",correlation_save_percent[0,1])
```

```python
#to enter all the derived data into a new csv file
header = "ID,Income,Mortgage,Utilities,Grocery,Transport,Insurance,Healthcare,Entertainment,Misc,Total Expenses,Savings,Save Percent"
np.savetxt('derived_house_finances.csv', finance, delimiter=',', header=header, comments='', fmt='%.2f')
print("Derived data has been saved to 'derived_house_finances.csv'")
```
