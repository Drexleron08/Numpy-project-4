```python
import numpy as np

# 1. Setup
num_houses = 50
filename = 'house_finances.csv'

# 2. Generate Unique 8-digit Property IDs
# Using randint with a range to ensure exactly 8 digits
property_ids = np.random.randint(10000000, 100000000, size=(num_houses, 1))

# 3. Generate Monthly Income 
monthly_income = np.random.randint(30000, 150000, size=(num_houses, 1))

# 4. Generate 8 Expense Categories 
# Categories: Mortgage, Utilities, Groceries, Transport, Insurance, Healthcare, Entertainment, Misc
expenses = np.random.randint(500, 30000, size=(num_houses, 8))

# 5. Combine all data into one matrix
data = np.hstack([property_ids, monthly_income, expenses])

# 6. Define Headers
header = "Property_ID,Monthly_Income,Mortgage,Utilities,Groceries,Transport,Insurance,Healthcare,Entertainment,Misc"

# 7. Save to CSV
np.savetxt(filename, data, delimiter=',', header=header, comments='', fmt='%d')

print(f"Successfully created '{filename}' with 500 records.")
```
